<h3>Importing Necessary Libraries</h3>

```python
import numpy as np
import pandas as pd
import re
```

<h3>Scraping Necessary Data</h3>

```python
df=pd.DataFrame({"State":['Andaman and Nicobar Islands', 'Lakshadweep', 'Chandigarh', 'Delhi', 
                          'Dadra and Nagar Haveli and Daman and Diu', 'Puducherry', 'Telangana'], 
                 "Latitude":['11.62°N', '10.56°N', '30.73°N', '28.70°N', '20.18°N', '11.94°N', '18.11°N'], 
                 "Longitude":['92.73°E', '72.64°E', '76.78°E', '77.10°E', '73.02°E', '79.81°N', '79.02°E']})
data1=pd.read_html('http://www.quickgs.com/latitudinal-and-longitudinal-extents-of-india-indian-states-and-cities/')
lat_long=pd.DataFrame(data1[0])
lat_long=lat_long.append(df, ignore_index=True)
lat_long
```

```python
data2=pd.read_html('https://en.wikipedia.org/wiki/COVID-19_pandemic_in_India#covid19-container', 
                   match="State/Union Territory")
covid=pd.DataFrame(data2[0])
covid
```

<h3>Data Cleaning</h3>

```python
lat_long['State'][9]='Ladakh'
lat_long['Latitude'][9]='34.21°N'
lat_long['Longitude'][9]='77.62°E'
lat_long=lat_long.append({'State':"Jammu and Kashmir",
                         'Latitude':'34.08°N',
                         'Longitude':'74.80°E'}, ignore_index=True)
lat_long
```

```python
def data_pre(cord):
    return cord[0:5]
lat_long['Latitude']  = lat_long['Latitude'].apply(data_pre).astype('float')
lat_long['Longitude'] = lat_long['Longitude'].apply(data_pre).astype('float')
lat_long
```

```python
covid.columns = ['State', 'Total Cases', 'Deaths', 'Recoveries', 'Active Cases', '1', '2', '3', '4']
covid.drop(columns=['1', '2', '3', '4'], inplace=True)
covid=covid.iloc[:-2, :]
covid
```

```python
covid['Recoveries'].astype('int')
covid['Active Cases'].astype('int')
count=0
for i in covid['Total Cases']:
    if not i.isdigit():
        i=re.sub("[^0-9]", "", i)
        covid['Total Cases'][count]=i
    count=count+1
covid['Total Cases'].astype('int')
count=0
for i in covid['Deaths']:
    if not i.isdigit():
        i=re.sub("[^0-9]", "", i)
        covid['Deaths'][count]=i
    count=count+1
covid['Deaths'].astype('int')
covid
```

```python
data=pd.merge(lat_long, covid, how='inner', on='State')
data.to_csv(r'covid-data.csv', index=False, header=True)
data
```
